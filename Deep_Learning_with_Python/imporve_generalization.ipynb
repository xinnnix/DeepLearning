```python
from tensorflow.keras.datasets import imdb
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
(train_data, train_labels), _ = imdb.load_data(num_words=10000)
```

```python
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results
```

```python
train_data = vectorize_sequences(train_data)
```

```python
model = keras.Sequential([
    layers.Dense(16, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history_original = model.fit(train_data, train_labels,
                             epochs=20, batch_size=512, validation_split=0.4)
```

```python
model = keras.Sequential([
    layers.Dense(4, activation='relu'),
    layers.Dense(4, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history_smaller_model = model.fit(train_data, train_labels,
                             epochs=20, batch_size=512, validation_split=0.4)
```

```python
val_loss_original = history_original.history['val_loss']
val_loss_smaller = history_smaller_model.history['val_loss']
epochs = range(1, len(val_loss_original)+1)
plt.plot(epochs, val_loss_original, 'r-', label='Original val loss')
plt.plot(epochs, val_loss_smaller, 'g-', label='Smaller model val loss')
plt.xlabel('Epochs')
plt.ylabel('Validation loss')
plt.legend()
plt.show()
```

```python
model = keras.Sequential([
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

history_larger_model = model.fit(
    train_data, train_labels,
    epochs=20, batch_size=512, validation_split=0.4)
```

```python
val_loss_original = history_original.history['val_loss']
val_loss_smaller = history_smaller_model.history['val_loss']
val_loss_larger = history_larger_model.history['val_loss']

epochs = range(1, len(val_loss_original)+1)
plt.plot(epochs, val_loss_original, 'r-', label='Original val loss')
plt.plot(epochs, val_loss_smaller, 'g-', label='Smaller model val loss')
plt.plot(epochs, val_loss_larger, 'b-', label='Larger model val loss')
plt.xlabel('Epochs')
plt.ylabel('Validation loss')
plt.legend()
plt.show()
```
